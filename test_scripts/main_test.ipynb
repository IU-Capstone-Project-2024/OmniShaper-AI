```python
from diffusers import StableDiffusionImageVariationPipeline, StableDiffusionPipeline
from PIL import Image
import torch
```

```python
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
prompt = "3D ultra realistic house, 8K Ultra HD, no backround, only house, many windows, aesthetic"
negative_prompt = "Full scale model, full scale model, bad quality, monotonous background"
image = pipe(prompt, negative_prompt=negative_prompt).images[0]

image.save("test.png")
```

```python
device = "cuda:0"
sd_pipe = StableDiffusionImageVariationPipeline.from_pretrained(
  "lambdalabs/sd-image-variations-diffusers",
  revision="v2.0",
  )
sd_pipe = sd_pipe.to(device)
```

```python
from torchvision import transforms
```

```python
im = Image.open("test.png")
tform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize(
        (224, 224),
        interpolation=transforms.InterpolationMode.BICUBIC,
        antialias=False,
        ),
    transforms.Normalize(
      [0.48145466, 0.4578275, 0.40821073],
      [0.26862954, 0.26130258, 0.27577711]),
])
inp = tform(im).to(device).unsqueeze(0)

out = sd_pipe(inp, guidance_scale=3)
out["images"][0].save("result.jpg")
```
